```python
#importando bibliotecas
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
```

```python
edp = pd.read_excel("C:/Users/gusta/Desktop/DNC/Learn_python/Dataset_colabs/clients_renamed.xlsx")
edp
```

```python
edp_group = edp.groupby(["AGE"]).mean().reset_index()
edp_group
```

```python
plt.figure(figsize=(10,8))
plt.plot(edp_group["AGE"], edp_group["valor_credito"])
plt.title('Credit by age')
plt.xlabel('AGE')
plt.ylabel('Credit')
plt.grid(b=True, color='gray', linestyle='--', linewidth=0.05)
```

```python
#seaborn
fig, axes = plt.subplots(2,2 ,figsize=(20,10))
sns.lineplot(data=edp, x='AGE', y='PAY_AMT2', hue="EDUCATION", style="EDUCATION", palette="Set2", ax=axes[0,0]).set(title ="Extrat in Setember 2005", xlabel=None);
sns.lineplot(data=edp, x='AGE', y='PAY_AMT3', hue="EDUCATION", style="EDUCATION", palette="Set2", ax=axes[0,1]).set(title ="Extrat in October 2005", xlabel=None);
sns.lineplot(data=edp, x='AGE', y='PAY_AMT4', hue="EDUCATION", style="EDUCATION", palette="Set2", ax=axes[1,0]).set(title ="Extrat in November 2005");
sns.lineplot(data=edp, x='AGE', y='PAY_AMT5', hue="EDUCATION", style="EDUCATION", palette="Set2", ax=axes[1,1]).set(title ="Extrat in Dezember 2005");
```

```python
fig = go.Figure()
from plotly.subplots import make_subplots

fig = make_subplots(rows=3, cols=2, shared_yaxes=True,
                    subplot_titles=("April","May","July","June","Augost","Setember"))
fig.add_trace(go.Scatter(x=edp_group["AGE"],
                         y=edp_group["PAY_AMT2"],
                         line=dict(color='red',width=1.5),
                         name="Payment send 04/2005"),row=1, col=1)
fig.add_trace(go.Scatter(x=edp_group["AGE"],
                         y=edp_group["PAY_AMT3"],
                         line=dict(color='purple',width=1.5),
                         name="Payment send 05/2005"),row=1, col=2)
fig.add_trace(go.Scatter(x=edp_group["AGE"],
                         y=edp_group["PAY_AMT4"],
                         line=dict(color='green',width=1.5),
                         name="Payment send 06/2005"),row=2, col=1)
fig.add_trace(go.Scatter(x=edp_group["AGE"],
                         y=edp_group["PAY_AMT5"],
                         line=dict(color='yellow',width=1.5),
                         name="Payment send 07/2005"),row=2, col=2)
fig.add_trace(go.Scatter(x=edp_group["AGE"],
                         y=edp_group["PAY_AMT6"],
                         line=dict(color='black',width=1.5),
                         name="Payment send 08/2005"),row=3, col=1)

fig.update_layout(title_text="Payment about months on 2005")
fig.update_yaxes(title_text="Payment value", titlefont_size=10)
fig.update_xaxes(title_text="AGE", row=2, col=2, titlefont_size=8)      
fig.update_xaxes(title_text="AGE", row=3, col=1, titlefont_size=8)  
fig.write_html("C:/Users/gusta/Desktop/DNC/Learn_python/Dataset_colabs/Payment_age.html")
```
