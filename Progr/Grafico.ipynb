```python
#declarando variaveis
nome = ['Lucas', 'Gustavo', 'Larissa', 'Paulo', 'Luis', 'Joel', 'Joelma']
idade = [24, 32, 36, 38, 35, 23, 26]
salario = [4000, 4500, 6000, 10000, 7000, 3000, 3500]
pessoas = []
c = []

#criando lista composta
for v1,v2,v3 in zip(nome, idade, salario):
    c.append(v1)
    c.append(v2)
    c.append(v3)
    pessoas.append(c[:])
    c.clear()

#imprimindo os nomes e idades
for p in pessoas:
    print(f' nome: {p[0]} idade: {p[1]} salario(desejÃ¡vel): {p[2]}')

#tirando media
media = sum(salario)/len(salario)
print('_'*50)
print(f'A mÃ©dia dos salÃ¡rios Ã©: {media:4.2f}')

#plotando graficos
plt.figure(figsize = (18,5))
plt.bar(nome, idade)
plt.title('Nome x Idade') 

#plotando graficos
plt.figure(figsize = (18,5))
plt.bar(nome, salario)
plt.title('Nome x Idade') 
```
