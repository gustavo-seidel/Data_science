```python
#importando bibliotecas
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
```

```python
edp = pd.read_excel("C:/Users/gusta/Desktop/DNC/Learn_python/Dataset_colabs/clients_renamed.xlsx")
edp
```

```python
edp_group = edp.groupby(["AGE"]).mean().reset_index()
edp_group
```

```python
plt.figure(figsize=(10,8))
plt.plot(edp_group["AGE"], edp_group["valor_credito"])
plt.title('Credit by age')
plt.xlabel('AGE')
plt.ylabel('Credit')
plt.grid(b=True, color='gray', linestyle='--', linewidth=0.05)
```

```python
#seaborn
fig, axes = plt.subplots(2,2 ,figsize=(20,10))
sns.lineplot(data=edp, x='AGE', y='PAY_AMT2', ax=axes[0,0]);
sns.lineplot(data=edp, x='AGE', y='PAY_AMT3', ax=axes[0,1]);
sns.lineplot(data=edp, x='AGE', y='PAY_AMT4', ax=axes[1,0]);
sns.lineplot(data=edp, x='AGE', y='PAY_AMT5', ax=axes[1,1]);
```
